## Sequential Ensemble (aka Boosting)

Machine learning ensemble meta-algorithm for principally reducing bias, and furthermore variance in supervised learning, and a group of machine learning algorithms that convert weak learner to string ones.

- Adaboost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_features=2, n_informative=2,
n_redundant=0,
random_state=0, shuffle=False)
```

```python
clf = AdaBoostClassifier(n_estimators=100, learning_rate = 0.5, random_state=0)

clf.fit(X, y)
clf.score(X, y)
```

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plot_decision_regions(X, y, clf=clf, legend=2)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('AdaBoost')
plt.show()
```

## Comparing Hyperparameters

```python
import itertools
import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(2, 2)

clf1 = AdaBoostClassifier(n_estimators=100, random_state=0)
clf2 = AdaBoostClassifier(n_estimators=10, random_state=0)
clf3 = AdaBoostClassifier(n_estimators=5, random_state=0)
clf4 = AdaBoostClassifier(n_estimators=1, random_state=0)

fig = plt.figure(figsize=(10,8))

labels = ["AdaBoost_100", "AdaBoost_10", "AdaBoost_5", "AdaBoost_1"]

for clf, lab, grd in zip([clf1, clf2, clf3, clf4],
                         labels,
                         itertools.product([0, 1], repeat=2)):

    clf.fit(X, y)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X, y=y, clf=clf, legend=2)
    plt.title(f"{lab}, {clf.score(X, y)}")

plt.show()
```

## Parallel Ensemble (aka Bagging)

Machine learning ensemble meta-algorithm intended to improve the strength and accuracy of machine learning algorithms used in classification and regression purpose. It additionally diminishes fluctuation of data(variance)and help to from over-fitting.

- Random Forest
- Extra Trees

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

clf = RandomForestClassifier(max_depth=5, random_state=0)
clf.fit(X, y)
```

```python
plt.figure(figsize=(8,8))
plot_decision_regions(X, y, clf=clf, legend=2)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Random Forest')
plt.show()
```

```python
gs = gridspec.GridSpec(2, 2)

clf1 = RandomForestClassifier(max_depth=20, random_state=0)
clf2 = RandomForestClassifier(max_depth=10, random_state=0)
clf3 = RandomForestClassifier(max_depth=5, random_state=0)
clf4 = RandomForestClassifier(max_depth=2, random_state=0)

fig = plt.figure(figsize=(10,8))

labels = ["RF_maxdepth_10", "RF_maxdepth_5", "RF_maxdepth_2", "RF_maxdepth_1"]

for clf, lab, grd in zip([clf1, clf2, clf3, clf4],
                         labels,
                         itertools.product([0, 1], repeat=2)):

    clf.fit(X, y)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X, y=y, clf=clf, legend=2)
    plt.title(f"{lab}, {clf.score(X, y)}")

plt.show()
```
