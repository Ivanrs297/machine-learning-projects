```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X, y, coef = make_regression(
    n_samples = 1000,
    coef = True,
    n_features = 1,
    noise = 10.5,
    bias = 10,
    random_state = 42
    )

print(coef)
plt.scatter(X, y)
plt.title("Data")
plt.xlabel("X")
plt.ylabel("y")
```

```python
df = pd.DataFrame(X)
df['target'] = y
df.to_csv("data.csv")
```

```python
class Model:
    def __init__(self, weight, bias):
        self.weight = np.array(weight)
        self.bias = bias
    
    def predict(self, X):
        return np.dot(X, self.weight) + self.bias


Model_A = Model(5, 2)
```

```python
y_pred = Model_A.predict(X)
```

```python
from sklearn.metrics import mean_squared_error

print(mean_squared_error(y, y_pred))

plt.scatter(X, y)
plt.plot(np.arange(-3, 5), Model_A.predict(np.arange(-3, 5)), c = "r")
```
