# Vector Norms

$L_p-norm$ $(p \geq 1)$ of a $d-dimensional$ vector:

$
    ||x||_p = \left( \sum_{i=1}^{d}|x_i|^{p} \right)^{\frac{1}{p}}
$

```python
from numpy import array
from numpy.linalg import norm

a = array([1, 2, 3, 4, 5])
print(a)
```

# $L_1$ norm of a vector
Is the sum of absolute values

$
    ||x||_1 = \sum_{i=1}^{d}|x_i|
$

```python
abs(a).sum()
```

```python
l1 = norm(a, 1)
print(l1)
```

# $L_2$ norm of a vector
Is the sum of squared values

$
    ||x||_2 = \left( \sum_{i=1}^{d}x_i^{2} \right)^{\frac{1}{2}} = \sqrt{\left( \sum_{i=1}^{d}x_i^{2} \right)}
$

```python
np.sqrt(np.square(a).sum())
```

```python
l2 = norm(a, 2)
print(l2)
```

# $L_\infty$ norm of a vector
Is the maximum norm, aka max norm

$
    ||x||_\infty =  \lim_{p \to \infty} \left( \sum_{i=1}^{d}|x_i|^{p} \right)^{\frac{1}{p}}
$

$
||x||_\infty = \max_{i}\{|x_i|; i = 1, 2, ..., d  \}
$

$
||x||_\infty = \max_{i}|x_i|
$

```python
from numpy import inf

maxnorm = norm(a, inf)
print(maxnorm)
```

# Plotting norms

```python
import pylab
from numpy import array, linalg, random, sqrt, inf

def plotUnitCircle(p):
    """ plot some 2D vectors with p-norm < 1 """
    for i in range(1000):
        x = array([random.rand()*2-1,random.rand()*2-1])
        if norm(x,p) < 1:
            pylab.plot(x[0],x[1],'bo')
            
    pylab.axis([-1.5, 1.5, -1.5, 1.5])
    pylab.show()
```

```python
plotUnitCircle(1), plotUnitCircle(2), plotUnitCircle(inf)
```
