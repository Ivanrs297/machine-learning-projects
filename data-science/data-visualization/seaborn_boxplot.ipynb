# Seaborn BoxPlot Demo

```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()

data = load_iris(as_frame=True)
df = data.data
df.head()
```

```python
plt.figure(figsize=(10,15))
boxplot = sns.boxplot(x="variable", y="value", data=pd.melt(df))
boxplot = sns.stripplot(x="variable", y="value", data=pd.melt(df), marker="o", alpha=0.2, color="black")
boxplot.axes.set_title("Iris BoxPlot", fontsize=16)
boxplot.set_xlabel("Features", fontsize=14)
boxplot.set_ylabel("Distribution", fontsize=14)
plt.show()
```
